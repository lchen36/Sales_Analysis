```python
df1 = pd.read_csv('feet.csv')
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2 = df1.dropna()
df2.isnull().sum()
```

```python
df2.head()
```

```python
 df2[df2['Order Date'].str[0:2] == 'Or']
```

```python
df3 =  df2[df2['Order Date'].str[0:2] != 'Or']
df3.head()
```

```python
df3['Order Date'] = pd.to_datetime(df3['Order Date'])
df3.head()
```

```python
df3['Quantity Ordered'] = pd.to_numeric(df3['Quantity Ordered'])
df3['Price Each'] = pd.to_numeric(df3['Price Each'])
```

```python
df3.head()
```

```python
df3['sales'] = df3['Quantity Ordered'] * df3['Price Each']
```

```python
df3.head()
```

```python
def get_month(x):
    return x.split('-')[1]
```

```python
df3['month'] = df3[df3['Order Date'].apply(lambda x: get_month(x))]
```
